```python
df.drop(columns=['EmployeeCount', 'Over18', 'StandardHours'], inplace=True)
```

```python
df.columns
```

```python
#create a column age_group
labels = ['Young Adult','Adult', 'Middle-aged', 'Senior']
df['Age_group'] = pd.qcut(df['Age'], q=4, labels=labels)
```

```python
df[['Age', 'Age_group']].head(15)
```

```python
#create a new column Education_level
education_labels = {
    1: 'Below College',
    2: 'College',
    3: 'Bachelor',
    4: 'Master',
    5: 'Doctor'
}
df['Education_level'] = df['Education'].map(education_labels)   
```

```python
df[['Education', 'Education_level']].head()
```

```python
#create a new column of environmentsatisfaction_level
EnvironmentSatisfaction_labels = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
df['EnvironmentSatisfaction_level'] = df['EnvironmentSatisfaction'].map(EnvironmentSatisfaction_labels)
```

```python
df[['EnvironmentSatisfaction', 'EnvironmentSatisfaction_level']].head()
```

```python
#create a new column Jobinvolvement_level
JobLevel_labels = {
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
}
df['JobInvolvement_level'] = df['JobInvolvement'].map(JobLevel_labels)
```

```python
df[['JobInvolvement', 'JobInvolvement_level']].head()
```

```python
#create a new column Jobsatisfaction_level
Jobsatisfaction_labels = {
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
}
df['JobSatisfaction_level'] = df['JobSatisfaction'].map(Jobsatisfaction_labels)
```

```python
df[['JobSatisfaction', 'JobSatisfaction_level']].head()
```

```python
#create a new column PerformanceRating_level
PerformanceRating_labels = {
1: 'Low',
2: 'Good',
3: 'Excellent',
4: 'Outstanding'
}
df['PerformanceRating_level'] = df['PerformanceRating'].map(PerformanceRating_labels)
```

```python
df[['PerformanceRating', 'PerformanceRating_level']].head()
```

```python
#create a new column RelationshipSatisfaction_level
RelationshipSatisfaction_labels = {
1: 'Low',
2: 'Medium',
3: 'High',
4: 'Very High'
}
df['RelationshipSatisfaction_level'] = df['RelationshipSatisfaction'].map(RelationshipSatisfaction_labels)
```

```python
df[['RelationshipSatisfaction', 'RelationshipSatisfaction_level']].head()
```

```python
#create a new column WorkLifeBalance_level
WorkLifeBalance_labels = {
1: 'Bad',
2: 'Good',
3: 'Better',
4: 'Best'
}
df['WorkLifeBalance_level'] = df['WorkLifeBalance'].map(WorkLifeBalance_labels)
```

```python
df[['WorkLifeBalance', 'WorkLifeBalance_level']].head()
```

```python
def salary_slab(income):
    if income < 5000:
        return "Low"
    elif income < 10000:
        return "Medium"
    elif income < 15000:
                return "High"
    else:
        return "Very High"

df["Salary_Slab"] = df["MonthlyIncome"].apply(salary_slab)
```
